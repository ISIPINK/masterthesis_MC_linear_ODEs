```python
import numpy as np
import matplotlib.pyplot as plt

qq = np.arange(0, 1, 0.001)

def f(x):
    return np.exp(1)*x - (x-1) 

uf = np.frompyfunc(f, 1, 1)

fig, ax = plt.subplots(figsize=(5, 6))  

ax.plot(qq, np.exp(qq) - (np.exp(1) - 1), label=r'$f$')
ax.plot(qq, uf(qq) - (np.exp(1) - 1), label=r'$\tilde{f}$')
ax.plot(qq, np.exp(qq) - uf(qq), label=r'$f-\tilde{f}$')
ax.set_xlabel("x")
ax.legend()
plt.rcParams.update({'font.size': 18})
plt.savefig("control_variate.png", dpi=300,bbox_inches='tight')
plt.show()
```
