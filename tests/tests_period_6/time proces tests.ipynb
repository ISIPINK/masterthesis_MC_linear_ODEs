In this notebook we do some tests to see how we can influence the time proces of RMC algorithm we want to get uniform distributed recursion calls.

```python
from random import random as U
from math import sqrt
import matplotlib.pyplot as plt


def time_dist(f:"fun"):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 6))
    T = [f(U()) for _ in range(10**3)]
    axes[0].hist(T)
    axes[0].set_title("base distribution")

    points = []
    def X(t):
        if U()<t: 
            S = f(U())*t
            points.append(S)
            X(S)
        

    for _ in range(10**3):
        X(1)
    axes[1].hist(points)
    axes[1].set_title("recursive distribution")
    plt.show()
```

```python
def f(x): return sqrt(x)
time_dist(f)
```
