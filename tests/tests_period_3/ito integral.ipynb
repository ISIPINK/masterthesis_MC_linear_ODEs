```python
import numpy as np 
from math import sqrt,  sin
import matplotlib.pyplot as plt

esol = []
nsim = 10**3
dt = 0.05
for _ in range(nsim):
    xs = np.arange(0,20,dt)
    sol = 0
    for x in xs: 
        sol += x*np.random.normal(loc=0.0, scale=1.0)*dt
    esol.append(sol)
    
plt.hist(esol, bins = 50)
print(np.average(esol))
plt.show()
    
```

```python
from random import random
from math import exp
import numpy as np 
from numpy.random import normal  
np.random.normal(loc=0.0, scale=1.0)

def N():
    return normal(loc=0.0, scale=1.0)
    
def S(t,dt,mu,sigma):
    return S(t-dt,dt,mu,sigma)*(1+ dt*(mu + sigma*N())) if dt<t else 1
    
tmp = []
nsim = 10**3
for _ in range(nsim):
    tmp.append(S(3,0.01,1,1))
    
plt.hist(tmp,bins=50)
plt.show()
print(np.average(tmp))
print(np.var(tmp))
 
```
