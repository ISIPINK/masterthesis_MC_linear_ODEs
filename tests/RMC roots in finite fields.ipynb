There some theory missing here. Let's solve
$$
x \equiv_5 x^2 +3
$$
this has solution 2 and 4 = -1 ?

```python
from random import randint, random
import matplotlib.pyplot as plt
def x():
    r = 2 if random()>0.3 else 3
    return (r* x() * x() +3)%5 if  randint(0,r-1)==0 else 3

l = []

for _ in range(100):
    res = 0
    for _  in range(1000):
        res += x()/1000
    l.append(round(res))
    
plt.hist(l)
plt.show()
```
