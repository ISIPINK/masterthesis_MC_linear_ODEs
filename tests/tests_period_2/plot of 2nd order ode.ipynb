This is an example out of period2.

```python
from random import random
import numpy as np
from math import exp,sqrt
import matplotlib.pyplot as plt

def G(t,x):
    return -t if t<x else -x 

memo = []
def Y(t):
    u = random()
    res = 1+ exp(1)*t +1.3*(Y(u)*G(t,u)) if random()<1/1.3 else 1+ exp(1)*t 
    memo.append((t,res))
    return res
    

sol = 0
nsim = 2*10**2
t = 0.7 # <1

for _ in range(nsim):
    sol += Y(t)/nsim

s = exp(t) #analytic solution
percentage_error = (sol - s)/s

print(f"sol({t}) is approx {sol},",f"%error = {percentage_error}")
xs,ys = zip(*memo)
zs = [exp(x) for x in np.arange(0,1,0.01)]
plt.scatter(xs,ys)
plt.plot( np.arange(0,1,0.01),zs, color="red")
plt.show()
plt.hist(xs,bins = 50)
plt.show()
```
